```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_corr = pd.read_parquet('correlation.park')
```

```python
df_corr
```

```python
def get_columns(df1_corr):
    s = abs(df1_corr).unstack()
    so = s.sort_values(kind="quicksort")

    features = (j[0] for i, j in zip(so[0.9 < so], so[0.9 < so].index) if j[0] != j[1])
    set_of_features = {i for i in features}

    return set_of_features

print(len(get_columns(df_corr)))
features_with_big_corr = list(get_columns(df_corr))
```

```python
df = pd.read_parquet('df_version_1.park')
print(df.shape)
df.drop(columns=features_with_big_corr, inplace=True)
df.shape
```
